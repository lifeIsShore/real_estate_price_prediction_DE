# Cleansing

```python
# comma to point
data_copy['airport_distance'] = data_copy['airport_distance'].str.replace(',', '.')

# Float
data_copy['airport_distance'] = data_copy['airport_distance'].astype(float)
```

```python
# comma to point
data_copy['city_score'] = data_copy['city_score'].str.replace(',', '.')

# Float
data_copy['city_score'] = data_copy['city_score'].astype(float)
```

```python
data_copy
```

```python
data_score_int = data_copy.copy()
```

all scores (geo-based) are handled and are turned into numeric values. and that is ,data_score_int, saved into df 

--------------------------------------------

```python
# take first part of price
data_score_int['price'] = data_score_int['price'].str.split(' ').str[0]

# make commas disapear and make them int
data_score_int['price'] = data_score_int['price'].str.replace(',', '')

# filter non int
data_score_int = data_score_int[data_score_int['price'].str.isdigit()]

data_score_int['price'] = data_score_int['price'].astype(float)
```

```python
# filter more then 3 million
data_score_int = data_score_int[data_score_int['price'] < 2_000_000]
```

```python
data_score_int.info()
```

```python
data_score_int
```

```python
data_price_clnsd = data_score_int.copy()
```

price column is trimed, and only included the first part (before the space charachter). and the price is filtered as no more record more than 2,000,000

---

```python
data_price_clnsd.info()
```

```python
# keep only the numerical
data_price_clnsd = data_price_clnsd[data_price_clnsd['number_of_rooms'].apply(lambda x: str(x).isdigit())]
# delete the row with 0
data_price_clnsd = data_price_clnsd[data_price_clnsd['number_of_rooms'].astype(int) != 0]

data_price_clnsd['number_of_rooms'] = data_price_clnsd['number_of_rooms'].astype(int)

data_price_clnsd = data_price_clnsd[data_price_clnsd['number_of_rooms'] < 11]
```

```python
# check the result
data_price_clnsd.info()
```

```python
data_price_clnsd
```

```python
data_rooms_clnsd = data_price_clnsd.copy()
```

rooms column is converted to int, and cleansed with all the str. and it is filtered from data that saying more 11 rooms 

---

```python
# Sayısal olmayan değerleri tespit et ve onları sil
data_rooms_clnsd['living_area'] = pd.to_numeric(data_rooms_clnsd['living_area'], errors='coerce')

# Sayısal olmayan (NaN) değerleri sil ve 30 dan kcükleri
data_rooms_clnsd = data_rooms_clnsd[(data_rooms_clnsd['living_area'] > 30) & (data_rooms_clnsd['living_area'].notna())]

# 'living_area' sütununu string'e dönüştür ve virgülleri noktaya çevir
data_rooms_clnsd['living_area'] = data_rooms_clnsd['living_area'].astype(str).str.replace(',', '.')

# Float türüne dönüştür
data_rooms_clnsd['living_area'] = data_rooms_clnsd['living_area'].astype(float)

# 500'den büyük değerleri sil
data_rooms_clnsd = data_rooms_clnsd[data_rooms_clnsd['living_area'] <= 500]

# Sonuçları kontrol et
print(data_rooms_clnsd['living_area'].describe())
```

```python
data_rooms_clnsd.info()
```

```python
data_living_size_clnsd = data_rooms_clnsd.copy()
```

str values are eliminated. nan and values less than 30 are also eliminated. , to . conversion is done. typecasting float  

---

```python
# convert into str
data_living_size_clnsd['land-size'] = data_living_size_clnsd['land-size'].astype(str)

# thousand delimiter -> the space char
#data_living_size_clnsd['land-size'] = data_living_size_clnsd['land-size'].str.replace('.', '', regex=False)

# decimal delimiter -> .
data_living_size_clnsd['land-size'] = data_living_size_clnsd['land-size'].str.replace(',', '.', regex=False)



#typecasting float
data_living_size_clnsd['land-size'] = pd.to_numeric(data_living_size_clnsd['land-size'], errors='coerce')
```

```python
data_living_size_clnsd['land-size'] = data_living_size_clnsd['land-size'].fillna(0)
```
